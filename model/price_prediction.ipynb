```python
import re
def numbers(string):
    pattern = r'(\d+(\.\d+)?)'
    match = re.search(pattern, string)
    return float(match.group(0))
```

```python
numbers("300Sq. Yards")
```

```python
def to_float(x):
    prices=x.split("-")
    if(len(prices)==2):
        avg=(float(prices[0].strip())+float(prices[1].strip()))/2
        return avg
    try:
        return float(x)
    except:
        if("Sq. Meter" in x):
            return numbers(x)*10.76
        if("Sq. Yards" in x):
            return numbers(x)*3
        if("Acres" in x):
            return numbers(x)*43560
        if("Guntha" in x):
            return numbers(x)*1089
        if("Grounds" in x):
            return numbers(x)*2400
        if("Perch" in x):
            return numbers(x)*272.5
        if("Cent" in x):
            return numbers(x)*435.6
        return None
```

```python
df["total_sqft"]=df["total_sqft"].apply(to_float)
```

```python
df["price_per_sqft"]=df["price"]*100000/df["total_sqft"]
df.head()
```
